```python
def signature(n):
    return ''.join(sorted(n))

print(signature('stop'))
```

```python
from pathlib import Path
file = Path("words.txt").read_text()
lines = file.splitlines()

scrabble = {}
def srabble_Descrambler():
    for line in lines:
        if (len(line) == 6):
            scrabble[signature(line)] = line

srabble_Descrambler()
print(scrabble)
```
